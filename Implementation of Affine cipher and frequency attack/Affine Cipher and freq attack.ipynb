```python
class AffineCipher:

    def cal_inv_mod(mode):
        InversesInMode ={}
        for i in range(mode):
            for j in range(mode):
                if (i * j) % mode == 1:
                    InversesInMode[i] = j
        return InversesInMode

    def decryption(cipher_txt, a_inve ,b, Inverses):
        plain_txt = ''
        for i in range(len(cipher_txt)):
            plain_txt += chr(((ord(cipher_txt[i]) - ord('A'))* a_inve - b) % 26 + ord('A'))
        return plain_txt

    def encryption(plain_txt, a ,b):
        cipher_txt = ''
        for i in range(len(plain_txt)):
            cipher_txt += chr(((ord(plain_txt[i]) - ord('A')) * a + b) % 26 + ord('A'))
        return cipher_txt
```

```python
freq_alpha = ['E', 'T', 'A', 'O', 'I', 'N', 'S', 'R', 'H', 'D', 'L', 'U', 'C', 'M', 'F', 'Y', 'W', 'G', 'P', 
        'B', 'V', 'K', 'X', 'Q', 'J', 'Z']
```

```python
class attcak:
    def cal_sorted_freq(cipher_txt):
        freq = {}
        for i in range(len(cipher_txt)):
            if cipher_txt[i] in freq:
                freq[cipher_txt[i]] += 1
            else:
                freq[cipher_txt[i]] = 1
        sorted_freq = sorted(freq.items(), key=lambda x: x[1], reverse=True)
        return sorted_freq  
    
    def freq_attack(cipher_txt, plain_txt , Inverses):
        sorted_freq = attcak.cal_sorted_freq(cipher_txt)
        for char1 in freq_alpha:
            for char2 in freq_alpha:
                for cipher_char1 in sorted_freq:
                    for cipher_char2 in sorted_freq:
                        if char1 == char2 or cipher_char1[0] == cipher_char2[0]:
                            continue

                        a = int( (ord(char1) - ord(char2)) / (ord(cipher_char1[0]) - ord(cipher_char2[0])) ) % 26
                        b = (ord(char1) - (a * ord(cipher_char1[0]))) % 26

                        if a in Inverses:
                            if AffineCipher.decryption(cipher_txt, a, b, Inverses) == plain_txt:
                                return a, b
        return -1,-1
```
