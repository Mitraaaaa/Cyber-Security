## Vigenere Impelementation

<div style="text-align: right; direction: rtl;">

1. **vigenere_encrypt(plain_text, key):**

 - هدف: رمزنگاری یک متن ورودی داده شده با استفاده از رمزنگاری ویجینر.
   - مراحل:
     - یک رشته خالی (encrypted_text) برای ذخیره متن رمزنگاری شده مقداردهی اولیه می‌شود.
     - کلید را تکرار می‌کند تا به طول متن ورودی برسد.
     - برای هر کاراکتر در متن ورودی و کاراکتر متن تکراری، اقدام به رمزنگاری هر کاراکتر الفبایی با استفاده از الگوریتم رمزنگاری ویژنر می‌شود.
     - کاراکترهای غیر الفبای بدون تغییر به رشته رمزنگاری شده اضافه می‌شوند.
     - متن رمزنگاری شده برگردانده می‌شود.


2. **vigenere_decrypt(encrypted_text, key):**

- هدف: رمزگشایی یک متن رمزنگاری شده داده شده با استفاده از رمزنگاری ویجینر.
  - مراحل:
    - یک رشته خالی (decrypted_text) برای ذخیره متن رمزگشایی شده مقداردهی اولیه می‌شود.
    - کلید را تکرار می‌کند تا به طول متن رمزنگاری شده برسد.
    - برای هر کاراکتر در متن رمزنگاری شده و کاراکتر متن تکراری، اقدام به رمزگشایی هر کاراکتر الفبایی با استفاده از الگوریتم رمزگشایی ویژنر می‌شود.
    - کاراکترهای غیر الفبای بدون تغییر به رشته رمزگشایی شده اضافه می‌شوند.
    - متن رمزگشایی شده برگردانده می‌شود.
</div>

```python
def vigenere_encrypt(plain_text, key):
    encrypted_text = ''
    key_repeated = (key * ((len(plain_text) + len(key) - 1) // len(key)))[:len(plain_text)]

    for char, key_char in zip(plain_text, key_repeated):
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            shift = ord(key_char.upper()) - ord('A')
            encrypted_text += chr((ord(char) - base + shift) % 26 + base)
        else:
            encrypted_text += char

    return encrypted_text

def vigenere_decrypt(encrypted_text, key):
    decrypted_text = ''
    key_repeated = (key * ((len(encrypted_text) + len(key) - 1) // len(key)))[:len(encrypted_text)]

    for char, key_char in zip(encrypted_text, key_repeated):
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            shift = ord(key_char.upper()) - ord('A')
            decrypted_text += chr((ord(char) - base - shift) % 26 + base)
        else:
            decrypted_text += char

    return decrypted_text
```

## Attack on Vigenere

### <span style="color: green; text-align: right; direction: rtl; font-weight: bold; float: right;"> توضیح توابع کلاسی که برای پیدا کردن تولید کلید از آن استفاده می‌کنیم.</span>

<div style="text-align: right; direction: rtl;">

1. **calculate_key_length(ciphertext, max_key_length=50):**
   - هدف: تعیین طول کلید احتمالی‌ترین استفاده شده در یک رمزگذاری جایگزینی.
   - مراحل:
     - یافتن دنباله‌های تکراری از حروف در متن رمزنگاری.
     - محاسبه فواصل بین موقعیت‌های زیرمتن‌های تکراری.
     - یافتن عوامل مشترک در فواصل محاسبه شده.
     - انتخاب طول کلید احتمالی بر اساس عوامل مشترک.
   - خروجی: طول کلید تخمین زده شده.

2. **find_repeated_substrings(ciphertext, length):**
   - هدف: یافتن زیرمتن‌های تکراری با طول مشخص در متن رمزنگاری.
   - مراحل:
     - حلقه زدهی از متن رمزنگاری برای یافتن زیرمتن‌های با طول مشخص.
     - حذف زیرمتن‌های غیر حروف الفبا.
     - ذخیره موقعیت زیرمتن‌های تکراری در یک دیکشنری.
   - خروجی: یک دیکشنری حاوی زیرمتن‌های تکراری و موقعیت آن‌ها.


3. **calculate_distances(substrings):**
  - هدف: محاسبه فواصل بین موقعیت‌های زیرمتن‌های تکراری.
   - مراحل:
     - گرفتن یک دیکشنری از زیرمتن‌های تکراری و موقعیت‌های آن‌ها.
     - محاسبه فواصل بین موقعیت‌های متوالی برای هر زیرمتن.
   - خروجی: یک لیست از فواصل بین موقعیت‌ها.


4. **find_common_factors(distances):**
   - هدف: یافتن عوامل مشترک در یک لیست از فواصل.
   - مراحل:
     - حلقه زدهی از فواصل و شناسایی عوامل تا یک حد مشخص.
     - استفاده از یک شمارنده برای شمارش تعداد هر عامل.
   - خروجی: یک شمارنده حاوی تعداد عوامل مشترک.


5. **choose_key_length(factors):**
   - هدف: انتخاب مشترک‌ترین عامل به عنوان طول کلید تخمین‌زده شده.
   - مراحل:
     - گرفتن یک شمارنده حاوی عوامل مشترک و تعداد آن‌ها.
     - انتخاب عامل با بیشترین تعداد به عنوان تخمین طول کلید احتمالی.
   - خروجی: طول کلید انتخاب‌شده.

</div>

```python
from collections import Counter

class Key_length:
        
    def calculate_key_length(ciphertext, max_key_length=50):
        distances = []
        for length in range(3, min(max_key_length, len(ciphertext) // 2)):
            substrings = Key_length.find_repeated_substrings(ciphertext, length)
            distances.extend(Key_length.calculate_distances(substrings))

        factors = Key_length.find_common_factors(distances)
        key_length = Key_length.choose_key_length(factors)

        return key_length

    def find_repeated_substrings(ciphertext, length):
        substrings = {}
        for i in range(len(ciphertext) - length + 1):
            substring = ciphertext[i:i + length]
            if substring.isalpha():
                if substring not in substrings:
                    substrings[substring] = [i]
                else:
                    substrings[substring].append(i)
        return {k: v for k, v in substrings.items() if len(v) > 1}

    def calculate_distances(substrings):
        distances = []
        for positions in substrings.values():
            distances.extend([positions[i + 1] - positions[i] for i in range(len(positions) - 1)])
        return distances

    def find_common_factors(distances):
        factors = Counter()
        for distance in distances:
            possible_factors = [factor for factor in range(2, min(distance, 20)) if distance % factor == 0]
            factors.update(possible_factors)
        return factors

    def choose_key_length(factors):
        # Choose the most common factor as the key length
        key_length = factors.most_common(1)[0][0]
        return key_length
```
